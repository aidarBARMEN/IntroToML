#### 6. Consistency of $\hat{\beta}$  

A sequence of estimators $\hat{\beta}_n$ is said to be **consistent** for $\beta$ if:  

$\hat{\beta}_n \xrightarrow{p} \beta \quad \text{(convergence in probability)}.$

This means that as the sample size increases, $\hat{\beta}_n$ gets arbitrarily close to $\beta$ with high probability.  

Using the **Law of Large Numbers (LLN)**, prove that the OLS estimator $\hat{\beta}$ is consistent under standard assumptions.  


**Proof of OLS Consistency TASK 6**:

We consider the linear model:
$$
y_i = x_i^\top \beta + \varepsilon_i,
$$
with i.i.d.\ $(x_i, \varepsilon_i)$. The OLS estimator is
$$
\hat{\beta}_n = (X_n^\top X_n)^{-1}X_n^\top y_n.
$$

Since $y_i = x_i^\top \beta + \varepsilon_i,$ we have
$$
\hat{\beta}_n 
= \beta 
+ 
(X_n^\top X_n)^{-1}X_n^\top \varepsilon_n.
$$

Under standard assumptions:
$$
\frac{1}{n} X_n^\top X_n 
\xrightarrow{p} 
\mathbb{E}[\,x_i x_i^\top\,]
\quad (\text{LLN}),
\qquad
\frac{1}{n} X_n^\top \varepsilon_n 
\xrightarrow{p} 
0.
$$

Thus,
$$
(X_n^\top X_n)^{-1} X_n^\top \varepsilon_n 
\xrightarrow{p} 
0,
\quad
\text{and}
\quad
\hat{\beta}_n 
\xrightarrow{p} 
\beta.
$$

Hence, OLS is consistent.

## Exercise 2 : Linear Dependence and Rank of a Matrix

**Understanding Linear Dependence in Matrices**

In the context of a matrix, **linear dependence** refers to a situation where one or more columns (or rows) can be written as a linear combination of other columns (or rows). If all columns (or rows) are linearly independent, no such relationship exists.

A set of column vectors $ \mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_n $ in $ \mathbb{R}^m $ is **linearly dependent** if there exist scalars $ c_1, c_2, \dots, c_n $, **not all zero**, such that:

$$
c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2 + \dots + c_n \mathbf{v}_n = \mathbf{0}
$$

Equivalently, a matrix $ A $ with columns $ \mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_n $ is **linearly dependent** if there is a nontrivial solution to:

$$
A \mathbf{c} = \mathbf{0}
$$

where $ A = [\mathbf{v}_1 \, \mathbf{v}_2 \, \dots \, \mathbf{v}_n] $ and $ \mathbf{c} \neq \mathbf{0} $.

One way to determine if a set of vectors (columns or rows) is linearly dependent is by checking the **rank** of the matrix.

- The **rank** of a matrix $ A $ is the number of **linearly independent** columns (or rows).
- If $ \text{rank}(A) = n $ (the number of columns), the columns are **linearly independent**.
- If $ \text{rank}(A) < n $, then at least one column can be written as a linear combination of the others, meaning the columns are **linearly dependent**.

To compute the rank of a matrix, we can transform it into **Row Echelon Form (REF)** using **Gaussian Elimination**. The number of nonzero rows in REF corresponds to the rank:

$$
\text{rank}(A) = \text{number of nonzero rows in REF}
$$

If $ \text{rank}(A) < n $, the matrix has **linearly dependent columns**.

```python
matrix1 = [

````python
def rank_matrix(matrix):
    #Create copy
    mat = [row[:] for row in matrix]
    rows = len(mat)
    cols = len(mat[0]) if rows > 0 else 0
    
    rank = 0
    pivot_row = 0
    
    #Walking via columns
    for col in range(cols):
        #looking for a row with a non-zero element in the current column (below pivot_row)
        pivot = None
        for r in range(pivot_row, rows):
            if mat[r][col] != 0:
                pivot = r
                break
        
        #If you haven't found it, go to the next column.
        if pivot is None:
            continue
        #Swapping the pivot row
        if pivot != pivot_row:
            mat[pivot_row], mat[pivot] = mat[pivot], mat[pivot_row]
        
        #Now we reset the column below pivot_row
        for r in range(pivot_row + 1, rows):
            if mat[r][col] != 0:
                #find the coefficient for subtraction
                factor = mat[r][col] / mat[pivot_row][col]
                # Вычтем factor * pivot_row из r-й строки
                for c in range(col, cols):
                    mat[r][c] -= factor * mat[pivot_row][c]
                    
        pivot_row += 1
        rank += 1      
        
        #If we run out of lines, exit
        if pivot_row == rows:
            break
    
    return rank

matrix1 = [[1, 2, 3], 
           [4, 5, 6], 
           [7, 8, 10]]
print("Rank of matrix1:", rank_matrix(matrix1))

matrix2 = [[1, 2, 3], 
           [2, 4, 6], 
           [3, 6, 9]]
print("Rank of matrix2:", rank_matrix(matrix2))

matrix3 = [
    [1, 2, 3, 4],
    [4, 5, 6, 7],
    [7, 8, 9, 10],
    [1, 1, 1, 1]
]
print("Rank of matrix3:", rank_matrix(matrix3))
````

### **Step 1**

You should implement a custom regression class using the template below that does the following:
- Parameters: Accepts a flag to include a constant (intercept) or not.
- Fit: Estimate the OLS parameters using the formulas
$$\widehat a = \frac{\sum\limits_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sum\limits_{i=1}^n (x_i - \bar{x})^2}, \quad 
    \widehat b = \bar{y} - \widehat a \bar{x}.$$
- Predict: Compute predictions using the estimated parameters.
- Residuals: Return the residuals, defined as:
$$ residuals = y − \widehat y $$

````python
import numpy as np

class CustomLinearRegression:
    def __init__(self, use_intercept=True):
        self.use_intercept = use_intercept
        self.coef_ = 0.0      # This will store the slope (a) of the regression line
        self.intercept_ = 0.0 # This will store the intercept (b), if enabled

    def fit(self, X, y):
        # Convert X and y to numpy arrays and make sure they are one-dimensional
        X = np.ravel(X)  # "Flattens" the array (if X was [n,1], it becomes [n,])
        y = np.ravel(y)

        if self.use_intercept:
            # Compute the means of X and y
            x_mean = np.mean(X)
            y_mean = np.mean(y)

            # Compute the numerator and denominator for the slope formula (a)
            numerator = np.sum((X - x_mean) * (y - y_mean))
            denominator = np.sum((X - x_mean)**2)

            # Calculate the estimated slope (a^) and intercept (b^)
            self.coef_ = numerator / denominator  # a^ (slope)
            self.intercept_ = y_mean - self.coef_ * x_mean # b^ (intercept)
        else:
            # If the intercept is not used, we assume b = 0
            # The formula for a simplifies to minimizing SSR through the origin
            numerator = np.sum(X * y)
            denominator = np.sum(X**2)

            self.coef_ = numerator / denominator
            self.intercept_ = 0.0  # Set b to 0 since we're forcing the line through (0,0)

    def predict(self, X):

        X = np.ravel(X)  # Ensure X is a 1D array
        return self.coef_ * X + self.intercept_  # Apply the linear equation

    def residuals(self, X, y):
        return y - self.predict(X)  # Calculate the difference between actual and predicted values
````

### Step 2

- Use your custom linear regression class to fit the model

````python
from sklearn.model_selection import train_test_split

# 1. Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,  #standart
    random_state=42 
)

# 2. Creating an instance of our custom regression model (including intercept)
model = CustomLinearRegression(use_intercept=True)

# 3. Training (fitting) the model on the training data
model.fit(X_train, y_train)

# 4. Making predictions on the test data
y_pred_test = model.predict(X_test)

# 5. Computing residuals (optional but useful for analysis)
resid_test = model.residuals(X_test, y_test)

# 6.Checking the learned parameters and the model's error
print("Slope (coef_):", model.coef_)
print("Intercept (intercept_):", model.intercept_)
print("Mean Squared Error (MSE) on test data:", 
      np.mean((y_test - y_pred_test)**2))

#Visualizing the results
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test, color='blue', label='Test data')  # Actual test data points
plt.scatter(X_test, y_pred_test, color='red', label='Predictions')  # Predicted points
plt.legend()
plt.show()
````

### Final Discussion

- Discuss how the linear model may underfit the data because it cannot capture the inherent non-linear relationships
- Explain how introducing polynomial terms leads to a better fit

Linear regression assumes that the relationship between feature x and response y can be described by an equation of the form "y=ax+b". 
This means that all data will be approximated in a straight line. If the actual dependence is clearly nonlinear (for example, a quadratic curve), then a simple straight 
line will not be able to capture the curve or complex shape of the graph. As a result, in some areas, the data will be systematically higher or lower than the models
predictions, and the residuals themselves (the difference between the actual and predicted values) will have a regular curved pattern instead of randomly fluctuating around zero.
This situation is called underfitting, because the model is too "simple" and does not "learn" the patterns in the data. In principle, it cannot represent a quadratic (or any other nonlinear) relationship, being 
limited by a straight line. As a result, the linear model will give afairly high error and characteristically structured residuals, indicating that the model does not have time to "repeat" the shape of the real graph.


When we include polynomial features (for example \(x^2, x^3\)) in a dataset, ordinary linear regression begins to look not only at \(x\) as a "linear" factor, but also at its degrees.

Instead of looking for a straight line \( y = w_0 + w_1x \), we now consider a generalized equation of the form:

$$
y = \overbrace{w_0 + w_1x + w_2x^2 + \dots + w_dx^d}^{\text{}}
$$
Although this is still formally "linear" regression in terms of parameters \( w_i \), the inclusion of powers of \( x \) allows the model to **capture and reproduce** nonlinearity in the data.

- If the true dependency follows the form \( x^2 \) (or something close to it), adding the \( x^2 \) term allows the model to adjust to a U-shaped (or more complex) curve.
- As a result, the **prediction graph** will no longer be a "straight line" but can take the shape of a polynomial of the required degree, more accurately capturing the data structure.
- This leads to a **reduction in systematic errors**: residuals lose their distinct nonlinear structure and become more random.
- Performance metrics (e.g., MSE, \( R^2 \)) usually improve significantly since the model eliminates typical *underfitting*, which occurs when trying to approximate a nonlinear dependency with a straight line.
